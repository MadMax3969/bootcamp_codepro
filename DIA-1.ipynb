```python
#Prepara alto sandwich de huevo

def Preparar_Sandwich():
    ingrediente_1 = 'Pan'
    ingrediente_2 = 'Huevos'
    ingrediente_3 = 'Mayonesa'
    ingrediente_4 = 'Tomate'

    print(f'Ya esta listo el Sandiwch que pediste, incluye: {ingrediente_1} {ingrediente_2} {ingrediente_3} {ingrediente_4}')


Preparar_Sandwich()

def Prepara_Tu_Propio_Sandwich():
    ingrediente_1 = input('Ingrese la clase de pan que desea')
    ingrediente_2 = input('Ingrese aderezo')
    ingrediente_3 = input('ingrese si desea carne pollo u mixto')
    ingrediente_4 = input('ingrese si desea tomate y lechuga')

    print(f'Wow preparaste tu propio Sandwich, te digo que incluye: Pan de {ingrediente_1}, con aderezo de {ingrediente_2}, es de {ingrediente_3} y tiene {ingrediente_4}')
          

Prepara_Tu_Propio_Sandwich()


#aqui donde es def, seriviria para definiri una funcion donde siempre tiene luego de los parentess :  y  cuando la llamas ya no necesitas de los : / 
```

```python
pedido_comida = input('Ingrese que plato desea')    #funcion global
pedido_bebida = input('Que bebida desea?')


def pedido_cliente(comida,bebida):   #parametros funcion personalizada local
    print(f'Preparando tu pedido de: {comida} con {bebida}') 
    print('Su pedido esta listo! y en camino 🛵')


pedido_cliente(pedido_comida,pedido_bebida)   #argumentos
```

```python
impuesto = 0.1   #variable global

def calcular_total(precio):
    total = precio * impuesto  #variable local dentro de la funcion
    print('El precio total es de:',total)

calcular_total(1000)
calcular_total(5000)
calcular_total(3000)
    
```

```python
# Challenge 🤺 - 20m

# Escribe un programa que pregunte al usuario su edad_usario para ingresar al festival Asunciónico y determine qué tipo de acceso puede tener.

# Si el usuario tiene menos de 18 imprimir: "Acceso denegado — Solo zona Kids 🎈"

# Si el usuario tiene mas de 18 pero menos de 21 imprimir: "Acceso permitido — No puede comprar bebidas alcoholicas 🎟️"

# Si el usuario tiene mas de 21 imprimir: "Acceso total — Disfruta el festival 🎉



edad_usario = int(input('Favor ingrese su edad'))

if  edad_usario < 18:
    print('Acceso Denegado - Solo Zona Kids 🎈')


elif edad_usario <= 21:
    print('Acceso permitido - No puede comprar bebidas alcoholicas 🎟️')


else:
    print('Acceso Total - Disfruta el festival 🎉')
```

```python
# Challenge 🤺 - 30m
# Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), imprima la temperatura escrita, y 
# luego clasificar la temperatura ingresada en los siguientes niveles:

# Entre -10 y 13 grados, que imprima "Ro'y koko".
# Entre 14 y 23 grados, que imprima "Refresca'i".
# Entre 24 y 28 grados, que imprima "Oikoite".
# Entre 29 y 45 grados, que imprima "Hakurooooo".
# Para todo lo demás, imprimir "Ñamano mba'e!".



temperatura_usuario = int(input("Cuantos grados decis que hace?"))
print('Segun vos hace', temperatura_usuario, '°C')


if temperatura_usuario >= -10 and temperatura_usuario <= 13:   #necesario para definir los limites
    print("Ro'y koko")


elif temperatura_usuario >= 14 and temperatura_usuario <= 23:
    print("Refresca'i")


elif temperatura_usuario >= 24 and temperatura_usuario <= 28:
    print("Oikoite")

elif temperatura_usuario >= 29 and temperatura_usuario <= 45:
    print("Hakurooooo")

else:
    print("Ñamano mba'e")
```

```python
# Challenge 🤺 - 20m
# Escribir un programa que almacene una contraseña en una variable, y pregunte al usuario por la contraseña hasta que introduzca la contraseña correcta.

# Cuando el usuario ingrese la contrasena correcta, imprimir un mensaje secreto.

contraseña_correcta = 'senna99425'


while True:
    contraseña_puesta = input('Favor ingrese su contraseña')
    if contraseña_puesta == contraseña_correcta:
        print('Welcome Stranger')
        break

    else:
        print('La contraseña es incorrecta')

```

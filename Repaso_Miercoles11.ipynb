```python
# Ejercicio (FM) 10m
# Objetivo:
# Pedir al usuario un número.
# El programa debe seguir pidiendo mientras el número sea distinto de 10.

# Cuando el usuario escriba 10, mostrar:
# "¡Correcto!"

contar = 0

while contar != 10:
    contar_ingresado = int(input('Adivine el numero:'))
    contar = contar_ingresado
if contar == 10:
    print('Correcto')
else:
    print('Siga intentando adivinar..')

print(contar)
```
